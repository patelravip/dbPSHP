```python
from __future__ import division
import os
import sys

from glob import glob
from random import shuffle
```

```python
genes = set(['gene_{:03d}'.format(x) for x in range(11)])
asFiles = set(['{}_AS.csv'.format(x) for x in genes])
```

```python
folders = glob('/DATA/raw/dbpshp/EP/*/')
print len(folders)
```

```python
folders[0]
```

```python
%%time
remaining = {}
done = []
for f in folders:
    asDone = os.path.join(f,'AS','*.csv')
    doneFiles = set(map(os.path.basename, glob(asDone)))
    notDone = list(asFiles-doneFiles)
    if len(notDone) == 0:
        done.append(os.path.basename(f.strip('/')))
    else:
        remaining[os.path.basename(f.rstrip('/'))] = notDone
```

```python
len(done)
```

```python
with open('/DATA/raw/dbpshp/ep_complete.txt', 'w') as fh:
    fh.write('\n'.join(done))
```

```python
MAO = '/home/tuf12855/RAW_DATA/EP/MAO/Living_seq_Nuc.mao'
```

```python
%%time
dataDir = '/home/tuf12855/DATA/dbpshp/{}'
cmds = []
for r,rg in remaining.items():
    dd = dataDir.format(r)
    for g in rg:
        seq = os.path.join(dd, 'seqs', g.replace('_AS.csv', '.fas'))
        nwk = os.path.join(dd, 'trees', g.replace('_AS.csv', '.nwk'))
        out = os.path.join(dd, 'AS', g.replace('.csv', ''))
        cmd = "megacc -a {mao} -d {fas} -t {tree} -o {out}".format(mao=MAO, fas=seq, tree=nwk, out=out)
        cmds.append(cmd)
```

```python
print len(cmds)
```

```python
shuffle(cmds)
with open('/DATA/raw/dbpshp/remaining_ep.txt', 'w') as fh:
    fh.write('\n'.join(cmds))
```
